```python
import pandas as pd
myData = pd.read_csv("twitter.csv")
myData.describe()
```

```python
myData.clase = pd.Categorical(myData.clase)
myData.clase = myData.clase.cat.codes
myData.describe()
#1 es la clase 'normal'
```

```python
normal = myData.loc[myData["clase"]==1].copy()
anomalo = myData.loc[myData["clase"]==0].copy()
normal.describe()
```

```python
anomalo.describe()
```

```python
import matplotlib.pyplot as plt
plt.boxplot([normal.intertime, anomalo.intertime], labels=["normal", "anomalo"])
plt.show()
```

```python
corr = myData.corr()
import seaborn as sns
sns.heatmap(corr)
```
