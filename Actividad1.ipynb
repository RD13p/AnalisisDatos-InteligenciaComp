```python
import pandas as pd
myData = pd.read_csv("iris.csv").drop(columns=['Id'],axis=1) #csv actividad 1
myData.describe()
```

```python
setosa = myData.loc[myData["Species"]=="Iris-setosa"].copy()
setosa.describe().loc[['mean', 'std']]
```

```python
versicolor = myData.loc[myData["Species"]=="Iris-versicolor"].copy()
versicolor.describe().loc[['mean', 'std']]
```

```python
virginica = myData.loc[myData["Species"]=="Iris-virginica"].copy()
virginica.describe().loc[['mean', 'std']]
```

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
scatter_matrix(myData)
plt.show()
```

```python
myData.Species = pd.Categorical(myData.Species)
myData.Species = myData.Species.cat.codes
corr = myData.corr()
import seaborn as sns
sns.heatmap(corr, annot=True)
```
